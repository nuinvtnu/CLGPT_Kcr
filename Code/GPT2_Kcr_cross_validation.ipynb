```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report
from sklearn.model_selection import KFold
from transformers import set_seed, GPT2Config, GPT2Tokenizer, GPT2ForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding

```

```python
path ="/content/drive/MyDrive/GPT2_Kcr/data/"
result_path ="/content/drive/MyDrive/GPT2_Kcr/Model/"
```

```python
# Định nghĩa đường dẫn dữ liệu
train_path = path + 'train_data.csv'
# Hàm chuyển đổi chuỗi thành K-mer
def k_mer_sentence(sequence, k):
    return " ".join(sequence[i:i + k] for i in range(len(sequence) - k + 1))
# Xử lý dữ liệu train
def process_train_data(train_path, k):
    df_train = pd.read_csv(train_path)
    df_train['n_gram'] = df_train['Sequence'].apply(lambda seq: k_mer_sentence(seq, k))
    X_train, y_train = df_train['n_gram'], df_train['Label']
    return X_train, y_train
# Thực hiện tiền xử lý với n-gram = 1
n = 1
X_train, y_train = process_train_data(train_path, n)
```

```python

class ProteinDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
# Sử dụng BertTokenizer để token hóa dữ liệu
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
# Kiểm thử chéo 10 mặt
set_seed(731)

# Khởi tạo tokenizer (không cần tạo lại mỗi fold)
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.padding_side = "left"
tokenizer.pad_token = tokenizer.eos_token

# Khởi tạo KFold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)
data = X_train
data_labels = y_train

# Danh sách lưu trữ kết quả
results = []
all_fprs, all_tprs, all_aucs = [], [], []
best_model = None
best_auc = 0
best_fold = None

plt.figure(figsize=(8, 6))  # Khởi tạo biểu đồ

# Vòng lặp qua từng fold
for fold, (train_index, test_index) in enumerate(kf.split(data), 1):
    print(f"\n Training Fold {fold}/{kf.get_n_splits()}")

    # Khởi tạo lại mô hình trong mỗi fold
    model_config = GPT2Config.from_pretrained('gpt2', num_labels=2)
    model = GPT2ForSequenceClassification.from_pretrained('gpt2', config=model_config)
    model.resize_token_embeddings(len(tokenizer))
    model.config.pad_token_id = model.config.eos_token_id

    # Chia dữ liệu huấn luyện và kiểm tra
    train_cv, test_cv = data.iloc[train_index].tolist(), data.iloc[test_index].tolist()
    train_cv_label, test_cv_label = data_labels.iloc[train_index].tolist(), data_labels.iloc[test_index].tolist()

    # Token hóa dữ liệu
    train_encodings_cv = tokenizer(train_cv, truncation=True, padding=True, max_length=512)
    test_encodings_cv = tokenizer(test_cv, truncation=True, padding=True, max_length=512)

    # Tạo Dataset
    train_cv1 = ProteinDataset(train_encodings_cv, train_cv_label)
    test_cv1 = ProteinDataset(test_encodings_cv, test_cv_label)

    # Cấu hình tham số huấn luyện
    training_args = TrainingArguments(
        output_dir=f'./results/fold_{fold}',
        num_train_epochs=5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        warmup_steps=500,
        weight_decay=0.01,
        logging_dir=f'./logs/fold_{fold}',
        logging_steps=10,
        learning_rate=4e-5,
    )

    # Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_cv1,
        eval_dataset=test_cv1,
        data_collator=DataCollatorWithPadding(tokenizer)
    )

    trainer.train()
    eval_results = trainer.evaluate()

    # Dự đoán
    preds_output = trainer.predict(test_cv1)
    preds = preds_output.predictions.argmax(-1)
    probas = preds_output.predictions[:, 1]

    # Confusion matrix
    TN, FP, FN, TP = confusion_matrix(test_cv_label, preds).ravel()

    # AUC
    fpr, tpr, _ = roc_curve(test_cv_label, probas)
    auc_score = auc(fpr, tpr)

    # Lưu kết quả
    results.append({'Fold': fold, 'TP': TP, 'FP': FP, 'TN': TN, 'FN': FN, 'AUC': auc_score})
    all_fprs.append(fpr)
    all_tprs.append(tpr)
    all_aucs.append(auc_score)

    if auc_score > best_auc:
        best_auc = auc_score
        best_model = model
        best_fold = fold

    print(classification_report(test_cv_label, preds, target_names=['Class 0', 'Class 1']))
    print(f"AUC Score: {auc_score:.4f}")

    plt.plot(fpr, tpr, label=f'Fold {fold} (AUC = {auc_score:.4f})')

# Vẽ ROC tổng hợp
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for 10-Fold Cross Validation of GPT-2')
plt.legend()
plt.grid()
plt.show()

# Lưu kết quả
output_file = os.path.join(result_path, '2.7.cv_CLGPT2_Kcr.xlsx')
df_results = pd.DataFrame(results)
df_results.to_excel(output_file, index=False)

# Lưu mô hình tốt nhất
best_model_path = os.path.join(result_path, f'2.7.Model_CL_GPT2_Kcr.pth')
torch.save(best_model.state_dict(), best_model_path)

print(f"\n Kết quả đã được lưu vào {output_file}")
print(f"Mô hình tốt nhất (Fold {best_fold}) với AUC = {best_auc:.4f} đã được lưu tại {best_model_path}")
```
